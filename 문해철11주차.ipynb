#### Dictionary
- dic.get() -> key 해당 value 반환
- dic.keys() -> 모든 키 반환
- dic.values() -> 모든 value 반환
- dic.items() -> 모든 key:value 튜플로 반환
- dic.pop(key) -> key 반환하고 값 반환
- dic.popitem() -> 마지막 key, value 제거
- dic.clear() -> 모든 항목 제거
- dic.find() -> key 에 해당하는 value 반환 key 없으면 -1 반환
- dic1.update(dic2) -> 딕셔너리 병합



```python
# dicionary key 는 string 선언해야됨 value 는 리스트랑 정수랑 상관 없는 듯
dic1 = {'a': 'ㅁ', 'b': 2, 'c': 3}
li = [1,2,3,4,5]
dic2 = {'li': li, 'e': 5, 'f': 6}
dic1.update(dic2)
print(dic1)
dic1.keys()
dic1.get('li')
```

#### Dictionary-Convinience Store practice

```python
item = {"coffee":1,"coke":1,"cup":1,"pen":1,"milk":1,"paper":1,"note":1}

while True:
    option = input("Please enter the manu 1) 재고 조회 2) 입고 3) 출고 4)종료: ")
    if option == "1":
        name = input("Enter the product name")
        print(f"{name} : {item[name]}")
    elif option == "2":
        name, count = input("Enter the product name: ").split()
        item[name] += int(count)
        print(f"{name} : {item[name]}")
    elif option == "3":
        name, count = input("Enter the product name: ").split()
        item[name] -= int(count)
        print(f"{name} : {item[name]}")
    elif option == "4":
        print("end")
        break
```

#### String.count() & List.count()
- String.count() -->> 스트링 내부의 특정 문자를 찾아줌
- List.count() -->> 리스트 내부의 특정 문자 또는 원소를 찾아줌 문자열일 경우 완벽히 일치 해야됨.

```python
a = ["banana", "a","b"]
c = ".".join(a)
print(f"기존 값: {a}\n기존 리스트의 데이터 타입: {type(a)}\n")
print(f"결과 값: {c}\n'.'.join(list)의 데이터 타입: {type(c)}\n")

li = list(range(0,10))
li.append("a")

print(f"a.count('a'): {a.count('a')}\nc.count('.'): {c.count('.')}")
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS # 조사나 # stopwords 를 제거
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
```

```python
a = open("/content/drive/Othercomputers/내 컴퓨터/Work Space/2024-2/문해철-실습/alice.txt").read()
a
wordcloud = WordCloud().generate(a)

wordcloud.words_ # 빈도수 출력

wordcloud2 = WordCloud(max_words= 100).generate(a) # 단어를 상위 빈도 100개만 자름
wordcloud2.words_
print(STOPWORDS)
print(type(STOPWORDS)) # set 은 Dictionary의 특수한 형태로 중복을 허용하지 않음
b= STOPWORDS.add("said") # set 임으로 add() method 사용
wordcloud3 = WordCloud(max_words = 100, stopwords=b).generate(a)
```

```python
plt.figure(figsize = (15,10)) # 그림 사이즈 15*10 으로 설정
plt.imshow(wordcloud, interpolation='bilinear') # wordcloud 객체를 imshow 함수에 넣으면 워드 클라우드 형태의 그래프 생성
plt.axis("off") # 눈금(격자) 지우기
plt.show()
```
