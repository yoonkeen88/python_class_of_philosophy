## logspace(시작 종료 갯수)

```python
arr = np.logspace(1,23,12)
## 같은 간격으로 일정한 갯수의 배열을 얻고 싶으면 이러헥 해라~
arr = arr.reshape(3,4)
print(arr)
arr = arr.reshape(4,-1) # 마이너스1 은 행과 열 하나만 해주면 나머지는 자동 계산
print(arr )
arr.flatten() # 평탄화 작업
```

## 난수 생성
- seed
- 실수는
  - rand(row,column)
- 정수는
  - randint(start, end, size = (row, column)


```python
np.random.seed(100)
print(np.random.rand(5,3))
print(np.random.random(4))
a,b=10,20
print(a*np.random.rand(10,3)+ a)
```

## 평균값과 정규분포



```python
np.random.seed(100)
m = 175
sd = 10
heights = m+sd*np.random.randn(10000) # 선형 변환 평균 + 편차*정규난수
print(heights)
mean = np.mean(heights)
med = np.median(heights)
print(mean, med)

```

```python
years = list(range(1950, 2011, 10) )

gdp = [67.0,80.0, 257.0,1686.0, 6505,11865.3, 22105.3]
plt.plot(years, gdp, color = "green", marker = "o", linestyle = "solid")
plt.title("Korean GDP by years")
plt.ylabel("$ - dallars")
plt.savefig("gdp_per_capita.png", dpi = 600)
plt.show()
```

```python
x = np.arange(-20, 20, 0.1)
y1 = x**2 +5
y2 = 2*x
y3 = -x**2 - 5

plt.plot(x,y1,"r-" ,x,y2,"g-", x,y3,"b--")
plt.axis([-30,30,-30,30])
plt.show()
```

- 원래 함수

```python
x = [x for x in range(-20,20,1)]
y1 = [i**2 +5 for i in x]
y2 = [i*2 for i in x]
y3 = [-i**2 - 5 for i in x]

plt.plot(x,y1, "r--", x,y2, "g^-", x,y3, "b*:")
plt.axis([-30,30,-30,30])
plt.show()
```

- 연습문제

```python
x = [i for i in range(-10,10)]
y = [2*i +2 for i in x]
plt.plot(x,y, marker = "o")
plt.axis([-10,10,-10,10])
plt.show()
```

- mathmatics

```python
x = [x for x in range(100)]
y = [y**2 for y in x]
z = [z**3 for z in x]
s = [2**z for z in x]

plt.plot(x,x, label = "linear")
plt.plot(x,y,  label = "quadratic")
plt.plot(x,z, label = "cubic")
plt.plot(x,s,  label = "exponential")

plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple Plot")
plt.axis([-100,100,-100,100])
plt.legend()
plt.show()
```

```python
##sin
x = np.arange(0,360);x
y = np.sin(np.radians(x))
y2 = np.cos(np.radians(x))
y3 = np.tan(np.radians(x))
plt.plot(x,y, label = "sin")
plt.plot(x,y2, "r", label = "cos")
# plt.plot(x,y3)
plt.legend()
plt.show()
```
