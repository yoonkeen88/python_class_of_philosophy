```python
def factorial_l(n):

  s = sum(range(n+1))
  return s

def factorial_a(n):

  s = sum(np.arange(1,n+1))

  return s

startl = time.time()
l = factorial_l(100)
endl = time.time()

start = time.time()
a = factorial_a(100)
end = time.time()
print(f"list 실행시간: {end-start}\n arr 실행시간: {endl-startl}",
      f"\nlist 실행 결과: {l}\n arr 실행 결과: {a}")
```

## 논리적인 인덱싱


```python
ages = np.array([18,19,25,30,28])
print(ages[1:3])
print(ages[ages>20]) ## 논리 연산자를 통한 출력
## 관계 식이 True 인 인덱스의 요소만 출력
print(ages<20)
```
